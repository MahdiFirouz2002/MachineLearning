```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the mean and standard deviation of the data
mean = 0
std_dev_low = 1
std_dev_high = 5

# Generate normally distributed data with the specified mean and standard deviation
low_std_data = np.random.normal(loc=mean, scale=std_dev_low, size=1000)
high_std_data = np.random.normal(loc=mean, scale=std_dev_high, size=1000)

# Calculate the standard deviation of the data
low_std = np.std(low_std_data)
high_std = np.std(high_std_data)

# Create a figure with two subplot and a twin subplot
fig, axs = plt.subplots(1, 2, figsize=(15, 7))
fig.subplots_adjust(wspace=0.4)
fig.suptitle("Histogram and Ideal Plot of Data Distribution")
ax1 = axs[0]
ax1_twinx = ax1.twinx()
ax2 = axs[1]
ax2_twinx = ax2.twinx()

# Plot the histogram on the subplots
ax1.hist(low_std_data, bins=30, edgecolor='black', alpha=0.5, label='Histogram')
ax2.hist(high_std_data, bins=30, edgecolor='black', alpha=0.5, label='Histogram')

# Plot the ideal plot on the second subplot for ax1
x_low = np.linspace(mean - 3 * std_dev_low, mean + 3 * std_dev_low, 1000)
pdf_low = 1 / (std_dev_low * np.sqrt(2 * np.pi)) * np.exp(-(x_low - mean)**2 / (2 * std_dev_low**2))
ax1_twinx.plot(x_low, pdf_low, color='blue', linewidth=3, label='Ideal Plot')

# Plot the ideal plot on the second subplot for ax2
x_high = np.linspace(mean - 3 * std_dev_high, mean + 3 * std_dev_high, 1000)
pdf_high = 1 / (std_dev_high * np.sqrt(2 * np.pi)) * np.exp(-(x_high - mean)**2 / (2 * std_dev_high**2))
ax2_twinx.plot(x_high, pdf_high, color='blue', linewidth=3, label='Ideal Plot')

# Set the plot title and labels for ax1
ax1.set_title(f"LOW_STD\n$\mu$={mean}, $\sigma$={std_dev_low}\nStandard Deviation: {low_std:.2f}")
ax1.set_xlabel('Data')
ax1.set_ylabel('Frequency')
ax1_twinx.set_ylabel('Probability Density')

# Set the plot title and labels for ax2
ax2.set_title(f"HIGH_STD\n$\mu$={mean}, $\sigma$={std_dev_high}\nStandard Deviation: {high_std:.2f}")
ax2.set_xlabel('Data')
ax2.set_ylabel('Frequency')
ax2_twinx.set_ylabel('Probability Density')

# Add a legend
ax1.legend(loc='upper left')
ax1_twinx.legend(loc='upper right')
ax2.legend(loc='upper left')
ax2_twinx.legend(loc='upper right')

# Display the plot
plt.show()
```

### Note:
as you can see the high standard deviation data is more spread out and has a wider range of values, while the low standard deviation data is more tightly clustered around the mean
