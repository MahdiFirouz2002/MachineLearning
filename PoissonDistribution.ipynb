```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

### Problem:
Suppose that on average, there are 5 cars passing through a busy intersection every minute. What is the probability that there will be 7 cars passing through the intersection in a given minute?

```python
# Set the expected value or mean
mu = 5
x_car = 7

# Define the x-axis values
x = np.arange(0, 15)

# Calculate the Poisson probabilities for each value of x
probabilities = poisson.pmf(x, mu)
prob = poisson.pmf(x_car, mu)

# print calculated probability
print("Probability of passing 7 car is: {:.2f}%".format(prob * 100))

# Plot the Poisson Distribution
plt.plot(x, probabilities, 'bo', ms=8)
plt.vlines(x, 0, probabilities, colors='b', lw=5)
plt.xlabel('Number of cars passing through intersection')
plt.ylabel('Probability')
plt.title('Poisson Distribution for Cars Passing Through Intersection')
plt.show()
```
