```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom
```

### Problem:
Suppose the probability of winning a prize in a lottery is 0.1. What is the probability that a person will have to buy at least 6 tickets to win a prize?


```python
# set the probability of success
p = 0.1
n = 6

# create an array of possible number of trials
n_trials = np.arange(1, 21)

# Calculate the probability of winning a prize on the nth ticket
prob = geom.pmf(n, p)

# Print the probability of winning a prize on the nth ticket
print("Probability of winning a prize on the {}th ticket: {:.2f}%".format(n, prob*100))

# calculate the Geometric probability mass function (pmf)
pmf = geom.pmf(n_trials, p)

# calculate the cumulative distribution function (cdf)
cdf = geom.cdf(n_trials, p)

# plot the Geometric pmf and cdf
fig, ax = plt.subplots(2, 1, figsize=(8, 8))

ax[0].stem(n_trials, pmf, use_line_collection=True)
ax[0].set_xlabel('Number of trials')
ax[0].set_ylabel('Probability')
ax[0].set_title('Geometric PMF (p = {})'.format(p))

ax[1].step(n_trials, cdf, where='post')
ax[1].set_xlabel('Number of trials')
ax[1].set_ylabel('Cumulative probability')
ax[1].set_title('Geometric CDF (p = {})'.format(p))

plt.tight_layout()
plt.show()
```
