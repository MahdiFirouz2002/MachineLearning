```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

# Problem:
A manufacturing process produces defective items with probability 0.2. 
If we randomly select 50 items from the process, what is the probability that 
less than 10 of them are defective?

```python
# Define the parameters of the Binomial Distribution
n = 50 # number of trials
p = 0.8 # probability of success

# Create an array of values for the random variable X
x = np.arange(0, n+1)

# Calculate the probability mass function of the Binomial Distribution
pmf = binom.pmf(x, n, p)

# Calculate the cumulative distribution function of the Binomial Distribution
cdf = binom.cdf(x, n, p)

# Calculate the probability of less than 10 defective items
less_than_10 = cdf[10]

# Print the result
print("Probability of less than 10 defective items:", less_than_10)

# Plot the probability mass function and cumulative distribution function
fig, ax = plt.subplots(2, 1, figsize=(8,8))

ax[0].bar(x, pmf)
ax[0].set_title('Binomial Distribution Probability Mass Function(PMF)')
ax[0].set_xlabel('Number of Successes')
ax[0].set_ylabel('Probability')

ax[1].step(x, cdf)
ax[1].set_title('Binomial Distribution Cumulative Distribution Function(CDF)')
ax[1].set_xlabel('Number of Successes')
ax[1].set_ylabel('Cumulative Probability')

plt.tight_layout()
plt.show()
```
